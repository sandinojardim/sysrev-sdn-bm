```python
import requests

# Define the API endpoint and API key
endpoint = "http://ieeexploreapi.ieee.org/api/v1/search/articles"
api_key = "cb2xk632ttfdccd7wkp75hs7"

# Define the search query and parameters
params = {
    "query": "benchmarking SDN controller",
    "apikey": api_key,
    "max_records": 200
}

# Send the API request and store the response
response = requests.get(endpoint, params=params)

# Check the status code of the response
if response.status_code != 200:
    raise Exception("Failed to retrieve results from IEEExplore API.")

# Convert the response to a JSON object
results = response.json()

print(results)
```
